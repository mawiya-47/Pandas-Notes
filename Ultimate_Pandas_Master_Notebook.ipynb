<h1 style="color:#fff;margin:0;font-size:30px;">🧵 Module 2 — Series</h1>
<p style="color:#e9d5ff;margin:6px 0 0 0;">The 1-dimensional labeled array — the building block of a DataFrame</p>
</div>

<div style="background:rgba(255,255,255,0.04);border-left:5px solid #7c3aed;
border-radius:10px;padding:16px 20px;margin:10px 0;">
<b style="color:#7c3aed;">📘 Definition</b>
<p style="color:#d1d5db;margin:8px 0 0 0;">A <b>Series</b> is a one-dimensional labeled array capable of holding any data type (integers, strings, floats, objects). Every value has an associated <b>index label</b>.</p>
</div>

```python
# Creating a Series from a list
s1 = pd.Series([10, 20, 30, 40], name="Scores")
print(s1)
print("\nIndex:", s1.index.tolist())
print("Values:", s1.values)
print("Name:", s1.name)
print("Dtype:", s1.dtype)
```

```python
# Custom index
s2 = pd.Series([85, 90, 78], index=["Ali", "Sara", "Zain"], name="Marks")
print(s2)

# Indexing & slicing
print("\nAli's mark:", s2["Ali"])
print("First two:\n", s2[:2])

# Vectorized operations
print("\nBonus applied:\n", s2 + 5)
```

<b style="color:#7c3aed;">📗 Definition</b>
<p style="color:#d1d5db;margin:8px 0 0 0;">A <b>DataFrame</b> is a 2D labeled data structure with columns of potentially different types — like a spreadsheet or SQL table, held in memory.</p>

```python
# From a dictionary
data = {
    "Name": ["Ali", "Sara", "Zain", "Hina"],
    "Age": [23, 25, 22, 24],
    "City": ["Karachi", "Lahore", "Islamabad", "Multan"]
}
df = pd.DataFrame(data)
df
```

```python
# From a list of lists
rows = [[1,"A"], [2,"B"], [3,"C"]]
df_list = pd.DataFrame(rows, columns=["id", "letter"])

# From a NumPy array
df_np = pd.DataFrame(np.random.randint(0,100,(3,3)), columns=["X","Y","Z"])

df_list, df_np
```

<b style="color:#7c3aed;">🗂️ Overview</b>
<p style="color:#d1d5db;margin:8px 0 0 0;">Pandas can read almost any file format: CSV, Excel, JSON, HTML tables, SQL databases, and pickle files — each via a dedicated <code>pd.read_*()</code> function.</p>

```python
# We'll create a small CSV in-memory to demonstrate read_csv()
from io import StringIO

csv_data = StringIO("""id,name,dept,salary
1,Ali,IT,85000
2,Sara,HR,62000
3,Zain,IT,91000
4,Hina,Finance,73000
""")
df_csv = pd.read_csv(csv_data)
df_csv
```

```python
# JSON example
json_data = """
[{"id":1,"product":"Laptop","price":1200},
 {"id":2,"product":"Mouse","price":25}]
"""
df_json = pd.read_json(StringIO(json_data))
df_json
```

```python
dirty = pd.DataFrame({
    "name": ["Ali","Sara",None,"Hina","Ali"],
    "score": [85, np.nan, 70, 95, 85],
    "grade": ["A","B","C","A+","A"]
})
dirty
```

```python
dirty["score"].fillna(dirty["score"].mean())
dirty.dropna()
```

```python
dirty.duplicated()
dirty.drop_duplicates()
dirty.rename(columns={"grade":"letter_grade"})
dirty.astype({"score": "float64"})
```

```python
df.groupby("dept").filter(lambda g: g["salary"].mean() > 70000)
```

### 🔀 GroupBy Workflow Diagram

```python
fig, ax = plt.subplots(figsize=(7,2.2), facecolor="#0f0c29")
ax.set_facecolor("#0f0c29"); ax.axis("off")
steps = ["SPLIT\n(by group)", "APPLY\n(function)", "COMBINE\n(result)"]
colors = ["#7c3aed","#ec4899","#22d3ee"]
for i,(s,c) in enumerate(zip(steps,colors)):
    ax.add_patch(plt.Rectangle((i*2.5,0),2,1, color=c, alpha=0.8))
    ax.text(i*2.5+1,0.5,s, ha="center", va="center", color="white", weight="bold")
    if i<2:
        ax.annotate("", xy=(i*2.5+2.3,0.5), xytext=(i*2.5+2,0.5),
                     arrowprops=dict(arrowstyle="->", color="white"))
ax.set_xlim(-0.3,6.8); ax.set_ylim(-0.3,1.3)
plt.show()
```

<h1 style="color:#fff;margin:0;font-size:30px;">🔗 Module 11 — Merge, Join & Concat</h1>
<p style="color:#e9d5ff;margin:6px 0 0 0;">Combining multiple DataFrames</p>
</div>

```python
left = pd.DataFrame({"id":[1,2,3], "name":["Ali","Sara","Zain"]})
right = pd.DataFrame({"id":[2,3,4], "dept":["IT","HR","Finance"]})

pd.merge(left, right, on="id", how="inner")
```

```python
pd.merge(left, right, on="id", how="left")
pd.merge(left, right, on="id", how="right")
pd.merge(left, right, on="id", how="outer")
```

```python
left.set_index("id").join(right.set_index("id"), how="inner")
```

```python
pd.concat([left, left], axis=0, ignore_index=True)
pd.concat([left, right], axis=1)
```

### 🎬 Project 1: Mini 'Netflix-style' Content Analysis

### 🚢 Project 2: Titanic-style Survival Analysis

<p style="color:#e5e7eb;"><b>Q: Why is my column showing dtype 'object' for numbers?</b><br>
A: It likely contains mixed types or strings — use <code>pd.to_numeric(col, errors='coerce')</code>.</p>
<p style="color:#e5e7eb;"><b>Q: Why did my in-place operation not work?</b><br>
A: Many Pandas methods return a new object by default; either reassign the result or pass <code>inplace=True</code>.</p>
<p style="color:#e5e7eb;"><b>Q: SettingWithCopyWarning — what does it mean?</b><br>
A: You modified a slice that pandas isn't sure is a view or a copy. Use <code>.loc[]</code> explicitly or call <code>.copy()</code> first.</p>
